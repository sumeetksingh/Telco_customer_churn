```python
td = pd.read_csv("/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
td.head()
```

```python
td.TotalCharges = pd.to_numeric(td.TotalCharges, errors = 'coerce')
td.info()
```

```python
td.TotalCharges.isnull().sum()
```

```python
td['TotalCharges'] = td['TotalCharges'].replace(" ",np.nan)
td = td[td['TotalCharges'].notnull()]
td.TotalCharges.isnull().sum()
```

```python
td['SeniorCitizen'] = td.SeniorCitizen.astype('object')

td.describe().round()
```

```python
df = td.copy()
```

```python
df.drop(["customerID"], axis=1, inplace=True)
```

```python
df['SeniorCitizen'].value_counts()
```

```python
df['Partner'].value_counts()
```

```python
df['Partner'] = df['Partner'].map(lambda s :1  if s =='Yes' else 0)
df['Partner'].value_counts()
```

```python
df['Dependents'] = df['Dependents'].map(lambda s :1  if s =='Yes' else 0)
df['PhoneService'] = df['PhoneService'].map(lambda s :1  if s =='Yes' else 0)
df['PaperlessBilling'] = df['PaperlessBilling'].map(lambda s :1  if s =='Yes' else 0)
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns = ['gender','MultipleLines','InternetService',
'OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaymentMethod','Churn'], drop_first = True)
```

```python

df['SeniorCitizen'] = df['SeniorCitizen'].map(lambda s :1  if s =='Yes' else 0)

df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors = 'coerce')
df.drop(['TotalCharges'], axis = 1, inplace = True)
```

```python
df['Churn_Yes'] = df['Churn_Yes'].astype(int)
Y = df["Churn_Yes"].values
X = df.drop(labels = ["Churn_Yes"],axis = 1)
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
weights = pd.Series(model.coef_[0],
 index=X.columns.values)
weights.sort_values(ascending = False)
```

All the customers that are likely to churn have positive values and customers unlikely to churn have negative values.
